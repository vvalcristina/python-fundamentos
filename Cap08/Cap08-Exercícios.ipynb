### Exercício 1

Crie um array NumPy com 1000000 e uma lista com 1000000.Multiplique cada elemento do array e da lista por 2 e calcule o tempo de execução com cada um dos objetos (use %time).
Qual objeto oferece melhor performance, array NumPy ou lista?

```python
ex_array = np.arange(1000000)
ex_list = list(range(1000000))
```

```python
# Tempo do array NumPy
%time for _ in range(10): ex_array_novo = ex_array * 2
```

```python
# Tempo list
%time for _ in range(10): ex_list_nova = [x * 2 for x in ex_list]
```

O array NumPy possui melhor performance, devido ao seu tempo de execução ser menor.

### Exercício 10

Crie um banco de dados no SQLite, crie uma tabela, insira registros, consulte a tabela e retorne os dados em dataframe do Pandas

```python
query = """
CREATE TABLE TABELA
(Nome VARCHAR(30),
Cargo VARCHAR(20),
Idade INTEGER
);"""

con = sqlite3.connect('dsa.db')
con.execute(query)
con.commit()
```

```python
data = [('Leia Organa', 'Senadora da República', 18),
       ('Darth Vader', 'General do Império', 40),
       ('Han Solo', 'Contrabandista', 25)]

insercao = "INSERT INTO TABELA VALUES(?,?,?)"

con.executemany(insercao,data)

con.commit()
```

```python
cursor = con.execute('select * from tabela')
rows = cursor.fetchall()
rows
```

```python
cursor.description
pd.DataFrame(rows, columns=[x[0] for x in cursor.description])
```
